```python
# importing libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# load data
df=pd.read_excel(r"C:\Users\mohan reddy\OneDrive\Desktop\self project\Superstore Dataset.xlsx")
df
```

```python
duplicate=df.duplicated().sum()
print(duplicate)
```

```python
df=df.drop_duplicates()
df
```

```python
#Total rows, columns, datatypes
df.shape
```

```python
#Convert Order Date / Return Date to datetime
df["Order Date"]=pd.to_datetime(df["Order Date"])
df["Ship Date"]=pd.to_datetime(df["Ship Date"])
df
```

```python
#Extract Year, Month, Day
df["year"]=df["Order Date"].dt.year
df["month"]=df["Order Date"].dt.month_name()
df["day"]=df["Order Date"].dt.day_name()
df
```

```python
#Delivery Days (Return Date – Order Date)
df["Delivery Days"] = (df["Ship Date"] - df["Order Date"]).dt.days
df
```

```python
# NumPy Calculations
print("max sales",np.max(df["Sales"]))
print("min sales",np.min(df["Sales"]))
print("Avg sales",np.mean(df["Sales"]))
```

```python
#Which product generated the highest sales?
df.groupby("Order ID")["Sales"].sum().sort_values
```

```python
#Which customer placed the most orders?
df["Customer Name"].value_counts()
```

```python
# Return vs Not Returned orders
df["return_status"]=df["Profit"].apply(lambda x: "Return" if x < 0 else "Not Return")
df
```

```python
# Return vs Not Returned orders
print(df['Return_Status'].value_counts())
```

```python
#. Total Refund / Loss due to returns
total_refund=df[df["Profit"]<0]["Profit"].sum()
print(total_refund)
```

```python
#Top 10 most profitable products
df.groupby("Order ID")["Profit"].sum().sort_values
```

```python
# Sales Trend 
sales=monthly_sales = df.groupby("month")["Sales"].sum()
print(sales)
```

```python
# saved cleaned set
df.to_csv("cleaned_ecommerce_dataset.csv", index=False)
print("\n CLEANED DATA SAVED AS CLEANED_ecommerce_dataset.csv")
```
